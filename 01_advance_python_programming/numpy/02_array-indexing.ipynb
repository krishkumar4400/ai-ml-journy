3. Boolean indexing

```python
arr = np.arange(11, 21)
arr
```

```python
# extract all even numbers
bool_index = arr % 2 == 0
bool_index
```

```python
bool_index.astype(int)
```

```python
bool_index
```

```python
arr = arr[bool_index]
arr
```

```python
arr
```
