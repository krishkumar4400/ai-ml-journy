Matrix operations

```python
mat1 = np.arange(1,31).reshape(6,5)
mat2 = np.arange(31,61).reshape(5,6)
```

```python
print(mat1)
print()
print(mat2)
```

```python
mat1 * mat2 # this is not actual matrix multiplication this is just element multiplication of both arrays
```

```python
# matrix multiplication
mat1 @ mat2
```

```python
A = np.array([[1,2],[3,4]])
A
```

```python
B = np.array([[5,6],[7,8]])
B
```

```python
print(A @ B)
# or 
print(np.dot(A,B))
```

```python
print(mat1.shape)
print(mat2.shape)
```

```python
mat1 = np.arange(1,31).reshape(6,5)
mat2 = np.arange(31,61).reshape(6,5)
```

```python
print(mat1)
print()
print(mat2)
```

```python
mat1.T @ mat2
```

```python
mat1 @ mat2.T
```

```python
# transpose
print(A)
print(A.T)
```

```python
# stacking - stack or merge two arrays
# we can perform stacking operation only in vectors or in one dimentional not in matrix
# we cannot stack 2 matrix because it is already stacked
```

```python
np.vstack((a,b))
```

```python
np.hstack((a,b))
```

```python
np.stack((a,b))
```

```python
np.column_stack((a,b))
```
