```python
import numpy as np
```

```python
s = np.array([
    [5,3,4,6,7,8,9,1,2],  # sum of a row must be 45 (1 + ... + 9 = 45) also for  column sum will be 45
    [6,7,2,1,9,5,3,4,8],
    [1,9,8,3,4,2,5,6,7],
    
    [8,5,9,7,6,1,4,2,3],
    [4,2,6,8,5,3,7,9,1],
    [7,1,3,9,2,4,8,5,6],
    
    [9,6,1,5,3,7,2,8,4],
    [2,8,7,4,1,9,6,3,5],
    [3,4,5,2,8,6,1,7,9] 
])

```

```python
s
```

```python
np.sum(s, axis=1)
```

```python
np.sum(s, axis=0)
```

```python
b = np.sum(s, axis=1)
b[b == 45]
b == 45
```

```python
b
```

```python
for i in b:
    if i != 45:
        print("Sudoku is not valid")
        break
else:
    print("sudoku is valid for rows")
```

```python
c = np.sum(s, axis=0)
for i in c:
    if i != 45:
        print("Sudoku is not valid")
        break
else:
    print("Sudoku is valid for columns")
```

```python
mat1 = s[0:3, 0:3]
mat1
```

```python
for i in range(0, 9, 3):
    for j in range(0, 9, 3):
        print(s[i:i+3, j:j+3])
        print(np.sum(s[i:i+3, j:j+3]))
```

```python
valid = True

for i in range(0, 9, 3):
    for j in range(0, 9, 3):
        if np.sum(s[i:i+3, j:j+3]) != 45:
            print("Sudoku is invalid")
            valid = False
            break
    if not valid:
        break

if valid:
    print("Sudoku is valid")

def check_sudoku(s):
    for i in range(0, 9, 3):
        for j in range(0, 9, 3):
            if np.sum(s[i:i+3, j:j+3]) != 45:
                return "Sudoku is invalid"
    return "Sudoku is valid"

print(check_sudoku(s))
```
