### Missing Data handling

- Finding Missing data
- Removing missing data
- filling the missing data

```python
import numpy as np 
import pandas as pd
```

Finding Missing Data

```python
data = {
    'A': [1,2,np.nan, 4,5],
    'B': [1, 2,3,4,5],
    'C': [1,2,3,np.nan, np.nan],
    'D': [1,np.nan, np.nan,np.nan,5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

Removing Missing Data

```python
df2 = df
df2
```

```python
df2.dropna()
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(thresh=3) # there should be atleast 3 non null value within a row
```

Filling the missing data

```python
df
```

```python
df.fillna(0, inplace=True) # fill null values with 0
```

```python
df
```

```python
# fill null values with different values

values = {
    'A': 0,
    'B': 100,
    'C': 200,
    'D': 300
}

df.fillna(value=values)
```

```python
df
```

```python
# fill nan with mean

df.fillna(df.mean())
```
