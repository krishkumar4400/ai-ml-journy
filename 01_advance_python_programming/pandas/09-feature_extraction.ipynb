```python
def extract_episode(txt):
    check = False
    data = ''
    for i in txt:
        if i== ")":
            check = False
            return data
        if check:
            data += i
        if i == '(':
            check = True
            
```

```python
for i in range(df.shape[0]):
    df["Episodes"] = extract_episode(df.loc[i]["Title"])
```

```python
df
```

```python
df["Episodes"] = df["Title"].apply(extract_episode)
```

```python
df['Episodes'] = df['Episodes'].str.replace(' eps', '')
```

```python
def extract_time(txt):
    data = ''
    
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range(i+1, i+20):
                data += txt[j]
    return data

```

```python
df['Timestamp'] = df['Title'].apply(extract_time)
```

```python
df
```

```python
# Split into start and end parts
df[["Start", "End"]] = df["Timestamp"].str.split(" - ", expand=True)

# Convert to datetime
df["Start"] = pd.to_datetime(df["Start"], format="%b %Y")
df["End"] = pd.to_datetime(df["End"], format="%b %Y")

# Calculate difference in months
df["Months"] = (df["End"].dt.year - df["Start"].dt.year) * 12 + (
    df["End"].dt.month - df["Start"].dt.month
)

print(df)
```
