```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x ** 2
y
```

```python
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.

plt.show()
```

```python
# multiple plots (subplot)

plt.subplot(2,2,1)
plt.plot(x+2, y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(x**2, x**3)
plt.subplot(2,2,2)
plt.plot(x**2, x**2)

plt.show()
```

```python
# object oriented way

fig = plt.figure()

axis1 = fig.add_axes([0.1, 0.1, 1, 1])
axis2 = fig.add_axes([0.3, 0.3, 0.8, 0.8])

axis1.plot(x, y)
axis2.plot(x, y)

plt.show()
```

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches

dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure()

axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis2 = fig.add_axes([0.6, 0.6, 0.3, 0.3])  # smaller inside plot

axis1.plot(x, y)
axis2.plot(x, y)

plt.show()
```

```python
fig = plt.figure(figsize=(12,8), dpi=100)

axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis2 = fig.add_axes([0.6, 0.6, 0.3, 0.3])  # smaller inside plot

axis1.plot(x, y)
axis2.plot(x, y)

plt.show()
```

```python
fig, axes = plt.subplots(1, 2)

axes[0].plot(x, y)
axes[1].plot(x, y)

plt.show()
```
