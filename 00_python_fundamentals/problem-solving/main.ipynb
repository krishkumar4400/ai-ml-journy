### Data Structures & Algorithms

- Lists, dicts, sets, tuples — know their time complexities

### OOP & Design Patterns

- Dunder/magic methods (__repr__, __enter__, __slots__, etc.)
- Metaclasses and descriptors
- SOLID principles in Python

## Intermediate Power Moves
### Functional Programming

- functools — lru_cache, partial, reduce
- itertools — chain, product, groupby
- Decorators (with arguments, stacked, class-based)

### Concurrency

- threading vs multiprocessing vs asyncio
- When to use each (I/O-bound vs CPU-bound)
- asyncio event loop and async/await

### Context Managers

- Writing your own with contextlib

## Advanced / Pro Level
### Performance

- Profiling with cProfile, line_profiler
- NumPy vectorization over loops
- Cython / Numba for hot paths
- __slots__ to reduce memory

### Internals

- CPython bytecode (dis module)
- The GIL — what it is and how to work around it
- Memory management and garbage collection

### Metaprogramming

- Metaclasses, __init_subclass__, type()
- Dynamic class creation
- Abstract Base Classes (abc)
