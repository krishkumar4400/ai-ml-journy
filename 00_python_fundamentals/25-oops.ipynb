## Constructor

__init__ Method

Constructor - sepcial method or functions that constructs or creates objects.

__init__ method - special method (constructor) that is used to initialize our object. it is a special method which gets called autmatically everytime we creates an object.
        if we don't write init method python will create it automatically

### Self parameter: 
it is storing the current instance of the class basically means it is refering to the current object. 

it is used when we want to refer properties and methods and attributes of the current object or we want to use properties of current object.

```python
import random


class Student:

    def __init__(self, name, email, password):  # parameterized constructor
        print("I am a constructor")
        # print(self)  # <__main__.Student object at 0x000002DE83A4E510>
        self.name = name
        self.email = email
        self.password = password

        id = random.random() + 1
        self._id = id

    # instance methods
    def get_id(self):
        return self._id


s1 = Student("krish kumar", "krish@gmail.com", "Krish@9661")
s2 = Student("ankit kumar", "ankit@gmail.com", "ankit@9661")
s3 = Student("raj kumar", "raj@gmail.com", "raj@9661")
print(s1.name)
print(s1.email)
print(s1.password)

print(s1.get_id())
print(s2.get_id())
print(s3.get_id())
```

## Attributes:
##
### There are 2 types of attributes in python:
#### 1. class attributes: attribute belong to class. -> common
#### 2. instance attributes: attribute which belong to object. -> unique

eg:
  #####  name, age, city -> instance attribute 
  #####  college_name -> class attribute

```python
class Student:
    college_name = "Shobhit University"  # class attributes
    PI = 3.1

    def __init__(self, name, age, city):
        self.name = name  # instance attributes
        self.age = age
        self.city = city
        self.PI = 3.14


s1 = Student("krish", 21, "Delhi")

print(s1.name)
print(s1.age)
print(s1.city)
print(s1.college_name)  # common  - no memory overhead

# print(Student.name)  # error -  type object 'Student' has no attribute 'name'

print(s1.college_name)

print(Student.college_name)

print(
    s1.PI
)  # 3.14 - instance value (instance attribute has higher priority than class attribute)
print(Student.PI)  # 3.1 # class attribute
```

```python
# instance method
class Laptop:
    storage_type = "SSD"

    def __init__(self, RAM, storage):
        self.RAM = RAM
        self.storage = storage

    def get_info(self):  # instance method
        print(f"Laptop has {self.RAM} RAM & {self.storage} {self.storage_type}")


l1 = Laptop("16GB", "512GB")
l2 = Laptop("8GB", "256GB")

l1.get_info()
l2.get_info()
```

```python
# class method
class Laptop:
    storage_type = "SSD"

    def __init__(self, RAM, storage):
        self.RAM = RAM
        self.storage = storage

    def get_info(self):  # instance method
        print(f"Laptop has {self.RAM} RAM and {self.storage} {self.storage_type}")

    @classmethod
    def get_storage_type(cls):  # class method
        print(cls.storage_type)


l1 = Laptop("32GB", "1TB")
l1.get_info()
l1.get_storage_type()
Laptop.get_storage_type()

# decorator - decorator is basically is a function that takes another function and chnages its behavior and return
```

```python
# static method
class Laptop:
    storage_type = "SSD"

    def __init__(self, RAM, storage):
        self.RAM = RAM
        self.storage = storage

    def get_info(self):  # instance attribute
        print(f"Laptop has {self.RAM} RAM and {self.storage} {self.storage_type}")

    @classmethod
    def get_storage_type(cls):
        print(f"The Laptop has {cls.storage_type} storage")

    @staticmethod
    def calculate_price(price, discount):
        print(f"Laptop price is {price - (price * discount / 100)}")


l1 = Laptop("64GB", "2TB")
l1.calculate_price(40_000, 10)
Laptop.calculate_price(50_000, 10)
```

## Problem - Product Store
#####
##### Desing & create an online store for produts (name, price)
##### Track total products being created.
##### Create a static method to calculate discount on each product based on a % parameter


```python
class Product:
    count = 0

    def __init__(self, name, price):
        self.name = name
        self.price = price
        Product.count += 1

    def get_info(self):  # instance method
        print(f"The Price of {self.name} is {self.price}")

    @classmethod
    def get_count(cls):
        print(f"Total Products in store = {cls.count}")

    @staticmethod
    def calculate_discount(price, discount):
        print(f"Discounted Price = {price - (price * discount / 100)}")


p1 = Product("Phone", 10_000)
p2 = Product("laptop", 40_000)
p3 = Product("pen", 10)
p4 = Product("book", 450)
p1.get_info()
print(p1.count)
print(p4.count)
print(Product.count)
p1.calculate_discount(10_000, 10)
Product.calculate_discount(p2.price, 22)
p4.calculate_discount(p4.price, 6)
Product.get_count()
```

## OOP Pillars
##### 1. Encapsulation: wrapping up of data and its related properties and attributes and methods in a single unit called class
    - in encapsulation we do also data hiding - data hiding basically means hiding sensitive or secure information from the user.

    => public data - accessible inside and outside everywhere in the class
    => protected data - accessible inside class and its related subclass
    => Private data - accessible inside the current class only.

##### 2. Abstraction: Hiding unneccessary details from the user and showing only necessary details.

##### 3. inheritance: Transfer properties of a class into its subclass or child class

##### 4. Polymorphism: use same functions and operators in multiple ways by changing its form.
